### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

In general the object-oriented approach is recommended because it is more flexible, reusable and allows for greater control. It is based on explicitly creating and managing Figures and Axes objects, which gives you a more fine tuned control over the layout and features of your plots. By calling methods on the plots you have direct interaction with the plot elements (e.g ax.plot() or ax.set_title()) rather than relying on global states. As such it is more scalable because the code is easier to structure for more complex visualizations or when creating plots in large applications. 

The Pyplot/MATLAB approach is instead a quick and interactive way to create plots. It is more implicit in the sense that when you call functions like plt.plot() or plt.xlabel() they apply to the current figure and axes automatically. Because of its state-based interferance and simpler syntax it is suitable and more convenient when you want to create a simple plot with a singel figure and axes. For example when doing exploratory data analysis where simplicity and speed matter more than flexibility.

### When plotting in Matplotlib, what is the expected input data type?

The expected data type when plotting functions in matplotlib is numpy.arrays, or objects that can be passed to the numpy.asarray in order to convert it to an numpy.array. For example a numpy.atrix might not work as intended and it is therefore recommended to convrt these to numpy.array objects prior to plotting.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Generate data
x = np.linspace(-4, 4)
y = x**2
```

```python
# Object-oriented approach
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Object-oriented plot')
ax.set_ylabel('y-values')
ax.set_xlabel('x-values')
```

```python
# Pyplot/MATLAB approach
plt.plot(x, y)
plt.ylabel('y-values')
plt.xlabel('x-values')
plt.title('Pyplot/MATLAB plot')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs = plt.subplots(2, figsize=(8, 12), layout='tight')

axs[0].scatter(random_data_x, random_data_y, facecolor='C0', edgecolor='k')
axs[0].set_title('Scatter plot', fontsize=16, color='orange')
axs[0].set_ylabel('Y-values')
axs[0].set_xlabel('x-values')

axs[1].bar(names, values, color=['purple', 'green', 'orange', 'yellow', 'lightgreen'])
axs[1].set_title('Bar plot', fontsize=16)
axs[1].set_ylabel('Amount')
axs[1].set_xlabel('Fruits', fontsize=14)

# Axis formatting.
axs[1].spines[['top', 'right', 'left']].set_visible(False)
axs[1].spines['bottom'].set_color('#DDDDDD')
axs[1].tick_params(bottom=False, left=False)
# Add horizontal grid
axs[1].set_axisbelow(True)
axs[1].yaxis.grid(True, color='#EEEEEE')
axs[1].xaxis.grid(False)
```
